```python
import numpy as np
import matplotlib.pylab as plt
import scipy.stats as stats
%matplotlib inline

np.seterr(all = 'ignore')
np.set_printoptions(precision=3, suppress = True)

plt.rc('text', usetex=True)
plt.rc('font',**{'family':'serif','serif':['Times New Roman']})
figSize  = (12, 6)
fontSize = 20
```

```python

x = np.array([53730.5293495834, 53555.7570653471, 53655.535759456, 53954.7396774769, 54534.4124955788, 56222.6934072569])
y = np.array([14.703633,14.74666, 14.979527, 14.912925, 14.945942, 14.482053])
yerr = np.array([0.0656027009062601, 0.0649612684378261, 0.0638530879223982, 0.0656591379435956, 0.0674724442918722, 0.0553135093155959])


data1 = np.linspace(y[0]-3*yerr[0],y[0]+3*yerr[0],10000)
data2 = np.linspace(y[1]-3*yerr[1],y[1]+3*yerr[1],10000)
data3 = np.linspace(y[2]-3*yerr[2],y[2]+3*yerr[2],10000)
data4 = np.linspace(y[3]-3*yerr[3],y[3]+3*yerr[3],10000)

fit1 = stats.norm.pdf(data1,y[0], yerr[0])
fit2 = stats.norm.pdf(data2,y[1], yerr[1])
fit3 = stats.norm.pdf(data3,y[2], yerr[2])
fit4 = stats.norm.pdf(data4,y[3], yerr[3])

choice1 = np.random.normal(y[0],yerr[0],1)
choice2 = np.random.normal(y[1],yerr[1],1)
choice3 = np.random.normal(y[2],yerr[2],1)
choice4 = np.random.normal(y[3],yerr[3],1)

plt.figure(figsize=figSize)
plt.scatter(x[0], y[0], c='black', marker='o', s = 30.0,lw=2.5)
plt.errorbar(x[0], y[0], yerr=yerr[0], linestyle='None',c='y',lw=2.5)
plt.plot(fit1+x[0],data1,'-',lw=2.5, c='b')
plt.hlines(y=choice1, xmin=min(fit1+x[0]), xmax=max(fit1+x[0])+5, color='r', linestyle='--',lw=2.5)

plt.scatter(x[1], y[1], c='black', marker='o', s = 30.0,lw=2.5)
plt.errorbar(x[1], y[1], yerr=yerr[1], linestyle='None',c='y',lw=2.5)
plt.hlines(y=choice2, xmin=min(fit2+x[1]), xmax=max(fit2+x[1])+5,color='r', linestyle='--',lw=2.5)
plt.plot(fit2+x[1],data2,'-',lw=2.5, c='b')

plt.scatter(x[2], y[2], c='black', marker='o', s = 30.0,lw=2.5)
plt.errorbar(x[2], y[2], yerr=yerr[2], linestyle='None',c='y',lw=2.5)
plt.plot(fit3+x[2],data3,'-',lw=2.5, c='b')
plt.hlines(y=choice3, xmin=min(fit3+x[2]), xmax=max(fit3+x[2])+5, color='r', linestyle='--',lw=2.5)


plt.xlabel("$Time \ (MJD)$", fontsize=fontSize)
plt.ylabel("$Magnitude$", fontsize=fontSize)
plt.xlim(53540,53750)
plt.tick_params(axis='both', which='major', labelsize=fontSize)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('./plots/RASLE.pdf')
plt.show()
```
