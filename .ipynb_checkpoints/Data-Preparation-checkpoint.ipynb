```python
def filename(path):
    # Load files
    files = glob.glob(path)

    n = 0
    file_name = []
    for file in files:
        filename = os.path.basename(str(file))
        file_name.append(filename[0:-13])
    
        n = n + 1
    print(np.shape(file_name))
    return file_name


def load_features(path, number_of_features):
    # Load files
    files = glob.glob(path)
    
    # compute the total number of files in the directory
    n = 0
    for file in files:
        n = n + 1
    
    # Create a matrix to store the original features with n_files for rows and n_features for columns
    master_matrix = np.zeros([n, number_of_features])
    
    rows = 0
    
    # Opening the data files containing the features
    for file in files:
        f    = open(file, 'r')
        data =  f.readlines()
        
        for i in range(number_of_features):
            master_matrix[rows][i] = float(data[i])
            
        rows =  rows + 1
        f.close()
        
    print("Master_matrix shape: %s" % str(master_matrix.shape))
    return master_matrix
```
