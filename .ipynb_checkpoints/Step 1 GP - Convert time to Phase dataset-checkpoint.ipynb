## Convert time series from time to phase domain

This script is used for Gaussian process. We need to convert all light curves from time to phase. We also apply sigma clipping pre-processing on the light curves.
Note: Correct index after sorting in ascending order so that it matches the correct magnitude and magnitude error.

```python
import numpy as np
import shutil
import os
import glob
import pandas as pd
import matplotlib.pylab as plt
#plt.switch_backend('agg')
% matplotlib inline

plt.rc('text', usetex=True)
plt.rc('font',**{'family':'serif','serif':['Palatino']})
figSize  = (12, 8)
fontSize = 20
```

```python
lC_dir  = './data/test_set/' # This needs to be changed either training or test set
savedir = './data/test_phase_lc/'
```

```python
info_data = pd.read_csv('./data/catalogue/SSS_Per_Tab.txt',delim_whitespace=True,names = ["Numerical_ID", "RA", "Dec", "Period", "V_CSS", "Npts", "V_amp", "Type", "Prior", "ID"])
info_data.head()
```

```python
period_data = pd.concat((info_data['Numerical_ID'], info_data['Period']), axis=1)
period_data.head()
```

```python
def sigma_clipping(date, mag, err, threshold=3, iteration=1):
    """
    Remove any fluctuated data points by magnitudes.
    
    Parameters
    ----------
    date      : an array of dates
    mag       : an array of magnitudes
    err       : an array of magnitude errors
    threshold : float, optional (Threshold for sigma-clipping) Here we use 3 sigma
    iteration : int, optional (the number of iteration)
    
    Returns
    -------
    date : an array of Sigma-clipped dates
    mag  : an array of Sigma-clipped magnitudes.
    err  : an array of Sigma-clipped magnitude errors.
    """

    if (len(date) != len(mag)) \
        or (len(date) != len(err)) \
        or (len(mag)  != len(err)):
        raise RuntimeError('Warning message: The length of date, mag, and err must be same.')

    # By magnitudes
    for i in range(int(iteration)):
        mean  = np.median(mag)
        std   = np.std(mag)
        index = (mag >= mean - threshold*std) & (mag <= mean + threshold*std)
        date  = date[index]
        mag   = mag[index]
        err   = err[index]

    return date, mag, err
```

The function below takes the light curves, apply a sigma clipping of 3 on it and then convert it into phase domain. It then saves it in .txt files for each light curves.

```python
file_list = os.listdir(lC_dir)
#file_list = file_list[0:5]
print(len(file_list))
for file in file_list:
    filename      = file[0:-4]
    row           = period_data[period_data.Numerical_ID.values.astype('float') == int(filename)]
    period        = float(row['Period'].values)
    data          = np.loadtxt(lC_dir + file)
    time_          = data[:,0]
    magnitude_     = data[:,1]
    magnitude_err_ = data[:,2]
    
    time_sigma, magnitude_sigma, mag_err_sigma=sigma_clipping(time_, magnitude_, magnitude_err_, threshold=3, iteration=1)
    # print('length of data is {} before clipping and length of data after clipping is {}'.format(len(magnitude_),len(magnitude_sigma)))
    
    phase_        = np.mod(time_sigma, period) / period
    i             = np.argsort(phase_)
    phase         = phase_[i]
    time          = time_sigma[i]
    magnitude     = magnitude_sigma[i]
    magnitude_err = mag_err_sigma[i]

    phase_df = pd.DataFrame(time)
    phase_df['Phase'] = phase
    phase_df['Magnitude'] = magnitude
    phase_df['Magnitude_Error'] = magnitude_err
    #phase_df.to_csv(savedir + filename +'.txt', sep=' ', index=None, header=None)
```

Check whether the light curve generated after tranformin from time to phase is correct

```python
file_list = os.listdir(savedir)
file_list = file_list[0:5]
for file in file_list:
    data          = np.loadtxt(savedir + file)
    phase         = data[:,1]
    magnitude     = data[:,2]
    magnitude_err = data[:,3]

    plt.figure(figsize=(12,8))
    plt.scatter(phase, magnitude, c='r', marker='o', s = 25.0)
    plt.errorbar(phase, magnitude, magnitude_err, linestyle='None',c='y')
    #plt.show()
    plt.close
```
