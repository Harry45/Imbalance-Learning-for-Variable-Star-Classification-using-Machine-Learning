```python
# RUN WITH PYTHON 2.7

import numpy as np
import shutil
import os
import glob
import pandas as pd
import matplotlib.pylab as plt
import scipy.optimize as op

from utils import sigma_clipping

from george import kernels
import george
#plt.switch_backend('agg')
% matplotlib inline

np.seterr(all = 'ignore')
np.set_printoptions(precision=3, suppress = True)

plt.rc('text', usetex=True)
plt.rc('font',**{'family':'serif','serif':['Times New Roman']})
figSize  = (12, 6)
fontSize = 20
```

```python
# Define the objective function (negative log-likelihood in this case).
def nll(p):
    gp.set_parameter_vector(p)
    ll = gp.log_likelihood(y, quiet=True)
    return -ll if np.isfinite(ll) else 1e25

# And the gradient of the objective function.
def grad_nll(p):
    gp.set_parameter_vector(p)
    return -gp.grad_log_likelihood(y, quiet=True)
```

```python
data_dir = './test-phase-light-curves/'
nSamples  = 3
files = glob.glob(data_dir+'*.txt')
labels = ['RRab', 'ACEP', 'Cep-II', 'RRc', 'RRd', 'Blazhko', 'Ecl', 'EA', 'Rot', 'LPV', '$\delta$-Scuti']
# print(files)
fig       = plt.figure(figsize=(24,14))
for i in range(0,len(files)):
    file_name = files[i][len(data_dir):-4]
    print(file_name)
    data   = np.loadtxt(files[i])

    
    phase    = data[:,0]
    mag      = data[:,1]
    mag_err = data[:,2]
            
    x, y, yerr = sigma_clipping(phase,mag,mag_err,threshold=3,iteration=2)
    

    kernel            = np.var(y) * kernels.Matern52Kernel(metric=0.1, ndim=1)
    gp                = george.GP(kernel,fit_white_noise=True)#,white_noise=np.log(0.19**2)
    comp              = gp.compute(x, yerr)
    covariance_matrix = gp.get_matrix(x)
    grad_likelihood   = gp.grad_log_likelihood(y, quiet=False)
    likelihood        = gp.log_likelihood(y, quiet=False)

    
    
    
    # Run the optimization routine.
    p0      = gp.get_parameter_vector()
    results = op.minimize(nll, p0, jac=grad_nll, method="L-BFGS-B")

    # Update the kernel and print the final log-likelihood.
    gp.set_parameter_vector(results.x)


    #Prediction: Calculate the mean and variance
    x_pred         = np.linspace(min(x), max(x), 1000)#x # The coordinates where the predictive distribution should be computed.
    pred, pred_var = gp.predict(y, x_pred, return_var=True)
    samples_lc     = gp.sample_conditional(y, x_pred, nSamples)

    index = i + 1
    ax    = plt.subplot(5,3,index)    
    plt.errorbar(x, y, yerr=yerr, fmt='o',ms =3,c="#DEB887",label=labels[i])##B0C4DE"
    y_error_neg = pred - 3*np.sqrt(pred_var)
    y_error_pos = pred + 3*np.sqrt(pred_var)
    plt.fill_between(x_pred, y_error_neg, y_error_pos,color='#CD950C', alpha=0.3)##104E8B
    plt.plot(x_pred, pred, "#8B2323", lw=2.0)
    plt.xlabel("$Phase$", fontsize=fontSize)
    plt.ylabel("$Mag$", fontsize=fontSize)
    plt.tick_params(axis='both', which='major', labelsize=fontSize)
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.20), prop={'size': 24})
    plt.xlim(-0.1,2.1)
    plt.gca().invert_yaxis()
    plt.tight_layout()

plt.subplots_adjust(hspace = 0.27)
plt.savefig('./plots/gp.pdf',bbox_inches = 'tight',pad_inches = 0.1)
plt.show()
```
