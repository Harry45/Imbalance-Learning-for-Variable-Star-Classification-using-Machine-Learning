```python
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import glob
import os

plt.rc('text', usetex=True)
plt.rc('font',**{'family':'DejaVu Sans','serif':['Palatino']})
figSize  = (12, 8)
fontSize = 20
```

```python
def sigma_clipping(date, mag, err, threshold=3, iteration=1):
    """
    Remove any fluctuated data points by magnitudes.
    Parameters
    ----------
    date : array_like
        An array of dates.
    mag : array_like
        An array of magnitudes.
    err : array_like
        An array of magnitude errors.
    threshold : float, optional
        Threshold for sigma-clipping.
    iteration : int, optional
        The number of iteration.
    Returns
    -------
    date : array_like
        Sigma-clipped dates.
    mag : array_like
        Sigma-clipped magnitudes.
    err : array_like
        Sigma-clipped magnitude errors.
    """

    # Check length.
    if (len(date) != len(mag)) \
        or (len(date) != len(err)) \
        or (len(mag) != len(err)):
        raise RuntimeError('The length of date, mag, and err must be same.')

    # By magnitudes
    for i in range(int(iteration)):
        mean = np.median(mag)
        std = np.std(mag)

        index = (mag >= mean - threshold*std) & (mag <= mean + threshold*std)
        date = date[index]
        mag = mag[index]
        err = err[index]

    return date, mag, err
```

```python
true_class=1
data_dir = '../Data/'
ascii_data  = pd.read_csv('../Ascii_SSS_Per_Table.txt',delim_whitespace=True,names = ["SSS_ID", "File_Name", "RA", "Dec", "Period", "V_CSS", "Npts", "V_amp", "Type", "Prior_ID", "No_Name1", 'No_Name2'])
types = ascii_data[ascii_data.Type==true_class]


files  = glob.glob(data_dir+'Type'+str(true_class)+'/*.dat')
files  = files[0:5]
print(len(files))

extracted_period = []
true_period = []
for file in files:
    filename = os.path.basename(str(file))
    file_name = filename[0:-4]

    crts_period = types[types.File_Name.values.astype(int)==int(file_name)].Period.values[0]
    crts_mean   = types[types.File_Name.values.astype(int)==int(file_name)].V_CSS.values[0]

    data    = np.loadtxt(file)
    date    = data[:,0]
    mag     = data[:,1]
    mag_err = data[:,2]


    time, magnitude, magnitude_err = sigma_clipping(date,mag,mag_err,threshold=3,iteration=2)
    print('length of data is {} before clipping and length of data after clipping is {}'.format(len(mag),len(magnitude)))
    N              = len(magnitude)

    snr                = np.sqrt((1/N)*np.sum(((magnitude - crts_mean)/magnitude_err)**2))
    half_frequency     = 1/crts_period
    uncertainty_freq   = half_frequency*np.sqrt(2/(N*snr**2))
    uncertainty_period = 1/uncertainty_freq

    print('The half-width at half-maximum is {} Â± {} days^-1'.format(half_frequency, uncertainty_freq))
    std_period = (crts_period**2)*uncertainty_freq
    
    print('The period is {} Â± {} days'.format(crts_period, std_period))
    
    mu = crts_period
    std = std_period
    num_samples = 10000
    
    new_period  = np.abs(np.random.normal(mu, std, num_samples))
    
    print('The sample period is {}'.format(new_period))
    
    plt.figure(figsize=figSize)
    plt.hist(new_period)
    plt.xlabel("Period in Days",fontsize=fontSize)
    plt.ylabel("Number of stars",fontsize=fontSize)
    plt.tick_params(axis='both', labelsize=fontSize)

    plt.show()
```
