```python
import numpy as np
import shutil
import os
import glob
import pandas as pd
import matplotlib.pylab as plt
import scipy.optimize as op

#plt.switch_backend('agg')
% matplotlib inline

np.seterr(all = 'ignore')
np.set_printoptions(precision=3, suppress = True)

plt.rc('text', usetex=True)
plt.rc('font',**{'family':'serif','serif':['Times New Roman']})
figSize  = (12, 12)
fontSize = 22
```

```python
level1_x    = np.array([9000, 11000, 13000, 15000, 17000])
level1_y    = np.array([78.3, 78.5, 78.7, 78.7, 79.0])
level1_yerr = np.array([0.005, 0.004, 0.004,0.005, 0.003])

level2i_x     = np.array([4000, 5000, 6000, 8000, 10000])
level2i_y     = np.array([92.9, 93.2, 93.1,93.2, 93.0])
level2i_yerr = np.array([0.004,0.005, 0.004, 0.005, 0.003])

level2ii_x    = np.array([9000, 11000, 13000, 15000, 17000])
level2ii_y    = np.array([99.3, 99.4, 99.4, 99.4, 99.4])
level2ii_yerr = np.array([0.001,0.003, 0.002, 0.002, 0.002])


level3i_x    = np.array([4000, 5000, 6000, 8000, 10000])
level3i_y    = np.array([89.0, 89.3, 89.5, 90.0, 90.2])
level3i_yerr = np.array([0.006, 0.005, 0.007, 0.004, 0.007])

level3ii_x    = np.array([1000, 2000, 3000, 4000, 5000])
level3ii_y    = np.array([86.3, 89.4, 90.1, 90.4, 91.1])
level3ii_yerr = np.array([0.05, 0.06, 0.07, 0.06, 0.08])

plt.figure(figsize = figSize)

plt.errorbar(level1_x, level1_y, yerr = level1_yerr*6,marker = '.',c='#6A5ACD',markersize=10, ecolor='#6A5ACD',capsize=4, elinewidth=3)
plt.plot(level1_x, level1_y, color = '#6A5ACD', lw = 2, label='Level 1')

plt.errorbar(level2i_x, level2i_y, yerr = level2i_yerr*6,c='#FF8247', marker = '.',markersize=10, ecolor='#FF8247',capsize=4, elinewidth=3)
plt.plot(level2i_x, level2i_y, color = '#FF8247', lw = 2, label='Level 2 Eclipsing Node')

plt.errorbar(level2ii_x, level2ii_y, yerr = level2ii_yerr*6,c='#006400',marker = '.',markersize=10, ecolor='#006400',capsize=4, elinewidth=3)
plt.plot(level2ii_x, level2ii_y, color = '#006400', lw = 2, label='Level 2 Pulsating Node')

plt.errorbar(level3i_x, level3i_y, yerr = level3i_yerr*6,c='#8B0000',marker = '.',markersize=10, ecolor='#8B0000',capsize=4, elinewidth=3)
plt.plot(level3i_x, level3i_y, color = '#8B0000', lw = 2, label='Level 3 RR Lyrae Node')

plt.errorbar(level3ii_x, level3ii_y, yerr = level3ii_yerr*6,c='#000080',marker = '.',markersize=10, ecolor='#000080',capsize=4, elinewidth=3)
plt.plot(level3ii_x, level3ii_y, color = '#000080', lw = 2, label='Level 3 Cepheids Node')
plt.xlim(800, 19000)

plt.ylabel(r"$Accuracy \ (\%)$ ",fontsize=fontSize)
plt.xlabel(r"$Number \ of \ training \ examples$ ",fontsize=fontSize)
plt.tick_params(axis='both', labelsize=fontSize)
plt.legend(loc='best', prop={'size': 18}, bbox_to_anchor=(1, 0.5))
plt.savefig('../plots/Effects_of_training_data.pdf',bbox_inches = 'tight',pad_inches = 0.1)
plt.show()

```
